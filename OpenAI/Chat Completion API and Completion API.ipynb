```python
import openai
import numpy as np

# Example texts
texts = [
    "OpenAI develops advanced AI models."

]

# Generate embeddings for each text
embeddings = [openai.Embedding.create(model="text-embedding-ada-002", input=text)['data'][0]['embedding'] for text in texts]

# Generate embedding for the query text
query_text = "OpenAI develops advanced AI models."
query_embedding = openai.Embedding.create(model="text-embedding-ada-002", input=query_text)['data'][0]['embedding']

# Function to calculate cosine similarity between two embeddings
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)

# Find the most similar text
similarities = [cosine_similarity(query_embedding, embedding) for embedding in embeddings]

# Get the index of the most similar text
most_similar_idx = np.argmax(similarities)


# Output the most similar text
print(f"The text most similar to '{query_text}' is:\n{texts[most_similar_idx]}")
```

```python
import openai



def get_completion(prompt,model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
    model=model,
    messages=messages,
    temperature=0,
    max_tokens = 100
    )

    return response.choices[0].message["content"]
```

```python
prompt = "How can I effectively prepare for my board exams?"
print(get_completion(prompt))
```
